# Distribution of ages:

An important aspect that we should tackle on is the distribution of ages of our audience, given the fact that we can target publicity towards that specific group.

```python
x_ = df1.member_birth_year.values
plt.hist(x = x_);
plt.title('Distribution of birth years')
plt.xlabel('years')
plt.ylabel('Amount of people');
```

```python
data_left = (df1.member_birth_year<1940).sum()
100 - data_left/df1.shape[0]*100
```

- Most of the data is concentrated after 1940. Numerically, around 99.9% of clients was borned after 1940.

```python
bins = np.arange(1940,df1.member_birth_year.max()+2,2)
birth_new = df1[df1.member_birth_year>=1940].member_birth_year.values
plt.hist(x = birth_new,bins = bins);
x_ticks = np.arange(1940,df1.member_birth_year.max()+5,5)
plt.xticks(x_ticks, x_ticks, rotation='vertical')

plt.title('Distribution of birth years between 1940 and 2005')
plt.xlabel('years')
plt.ylabel('Amount of people');
```

- The graph is left skewed with maximum between 1985 and 1990

# Distribution of start and end point using latitude and longitude

We want to know what are the most common places where people tend to use our service. So that we can attend that demand and transform it into profit

```python
plt.figure(figsize= [15,6])

plt.subplot(1,2,1)

lat = df2.start_station_latitude.values
long = df2.start_station_longitude.values

plt.hist2d(x=lat,y = long, cmap = 'viridis_r', cmin = 0.5);
plt.colorbar();
plt.title('Start locations',fontsize = 15)
plt.xlabel('latitude',fontsize = 10)
plt.ylabel('longitude',fontsize = 10)

plt.subplot(1,2,2)

plt.title('End locations',fontsize = 15)
plt.xlabel('latitude',fontsize = 10)
plt.ylabel('longitude',fontsize = 10)

end_lat = df2.end_station_latitude
end_long = df2.end_station_longitude

plt.hist2d(x=end_lat,y = end_long, cmap = 'viridis_r', cmin = 0.5);
plt.colorbar();
```

- The start and endpoint are almost identical, this could be due to the fact that the amount of data is big.
- There are two big areas of concentration.
- There is a point of maximum concentration around (37.78, -122.40)

# Amount of time per place
As we saw previously, there is a distribution with two main places. But, how about are this distributions related to the time that people ride the bike?

```python
import seaborn as sbn
plt.figure(figsize = (10,10))
plt.scatter(data = df3, x = 'start_station_latitude', y = 'start_station_longitude', c = 'duration_sec',
           cmap = 'mako_r')
plt.colorbar()
plt.title('Distribution of time consumed in the map',fontsize = 20)
plt.xlabel('latitude',fontsize=15)
plt.ylabel('longitude',fontsize = 15);
```

- The group located in the bottom part of the screen has more time consumed
- The distribution of points in the bottom part is more concentrated, the top group is more spread.
